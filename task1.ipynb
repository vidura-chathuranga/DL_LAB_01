```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
exponential_array = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Array from Exponential Distribution:")
print(exponential_array)
```

```python
exponential_array_large = np.random.exponential(scale=1.0, size=(100000, 1))
```

```python
# Create a random 100000x1 array from a uniform distribution
uniform_array_large = np.random.uniform(low=0.0, high=1.0, size=(100000, 1))

# Create a random 100000x1 array from a normal distribution
normal_array_large = np.random.normal(loc=0.0, scale=1.0, size=(100000, 1))
```

```python
plt.figure(figsize=(10, 6))

plt.hist(exponential_array_large, bins=100, alpha=0.5, label='Exponential', color='red')

plt.hist(uniform_array_large, bins=100, alpha=0.5, label='Uniform', color='blue')

plt.hist(normal_array_large, bins=100, alpha=0.5, label='Normal', color='green')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histograms of Exponential, Uniform, and Normal Distributions')
plt.legend()
plt.grid(True)
plt.show()
```
